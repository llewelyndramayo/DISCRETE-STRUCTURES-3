1. Without the use of any python libraries or modules, develop a function that can perform matrix addition, given two numpy matrices.

```python
def matrix_addition(mat1, mat2):
    if len(mat1) != len(mat2) or len(mat1[0]) != len(mat2[0]):
        raise ValueError("Matrices must have the same dimensions for addition.")

    result = []

    for i in range(len(mat1)):
        row = []
        for j in range(len(mat1[0])):
            row.append(mat1[i][j] + mat2[i][j])
        result.append(row)

    return result

# Example matrices
matrix1 = [[1, 1, 5], [2, 0, 3]]
matrix2 = [[2, 1, 3], [7, 1, 4]]

# Perform matrix addition
result_matrix = matrix_addition(matrix1, matrix2)

# Display the result
print("Matrix 1:")
for row in matrix1:
    print(row)

print("\nMatrix 2:")
for row in matrix2:
    print(row)

print("\nResult Matrix:")
for row in result_matrix:
    print(row)
```

2. Without the use of any python libraries or modules, develop a function that can multiply two numpy matrices.

```python
def matrix_multiplication(mat1, mat2):
    if len(mat1[0]) != len(mat2):
        raise ValueError("Incompatible matrices for multiplication.")

    result = [[0 for _ in range(len(mat2[0]))] for _ in range(len(mat1))]

    for i in range(len(mat1)):
        for j in range(len(mat2[0])):
            result[i][j] = sum(mat1[i][k] * mat2[k][j] for k in range(len(mat1[0])))

    return result

# Example matrices
matrix1 = [[1, 2, 3], [4, 5, 6]]
matrix2 = [[7, 8], [9, 10], [11, 12]]

# Perform matrix multiplication
result_matrix = matrix_multiplication(matrix1, matrix2)

# Display the result
print("Matrix 1:")
for row in matrix1:
    print(row)

print("\nMatrix 2:")
for row in matrix2:
    print(row)

print("\nResult Matrix:")
for row in result_matrix:
    print(row)
```

```python
def scalar_mult_add(k, mat1, mat2):
    result = []
    for i in range(len(mat1)):
        row = []
        for j in range(len(mat1[0])):
            row.append(k * (mat1[i][j] + mat2[i][j]))
        result.append(row)
    return result

def scalar_mult_then_add(k, mat1, mat2):
    result_A = []
    result_B = []

    for i in range(len(mat1)):
        row_A = []
        row_B = []
        for j in range(len(mat1[0])):
            row_A.append(k * mat1[i][j])
            row_B.append(k * mat2[i][j])
        result_A.append(row_A)
        result_B.append(row_B)

    return result_A, result_B


# Sample matrices
matrix_A = [[3, 1], 
            [5, 2]]
matrix_B = [[8, 4], 
            [1, 7]]
k = 2

result = scalar_mult_add(k, matrix_A, matrix_B)
result_A, result_B = scalar_mult_then_add(k, matrix_A, matrix_B)

# Display results row by row
print("Matrix A:")
for row in matrix_A:
    print(row)

print("\nMatrix B:")
for row in matrix_B:
    print(row)

print("\nk(A + B):")
for row in result:
    print(row)

print("\nkA + kB:")
for row in result_A:
    print(row)
for row in result_B:
    print(row)
```

4. Without using a python library or modules develop a function that can extract the diagonal of a numpy matrix.

```python
def extract_diagonal(matrix):
    if not matrix:
        return None  # Handle empty matrix case

    rows, cols = len(matrix), len(matrix[0])
    diagonal = []

    # Check if the matrix is square
    if rows != cols:
        return None  # Diagonal extraction is not possible for non-square matrices

    for i in range(rows):
        for j in range(cols):
            if i == j:
                diagonal.append(matrix[i][j])

    return diagonal

# Example matrix
example_matrix = [
    [5, 7, 1],
    [8, 1, 5],
    [1, 2, 9]
]

# Extract diagonal
result_diagonal = extract_diagonal(example_matrix)

# Display result
print("Original Matrix:")
for row in example_matrix:
    print(row)

print("\nExtracted Diagonal:")
print(result_diagonal)
```

5. Without using a python library or modules develop a function that can find a trace of a numpy matrix.

```python
def matrix_trace(matrix):
    if not matrix or len(matrix) != len(matrix[0]):
        return None  # Return None for empty or non-square matrices

    trace_sum = 0
    for i in range(len(matrix)):
        trace_sum += matrix[i][i]

    return trace_sum

# Sample matrix
example_matrix = [
    [5, 9, 6],
    [4, 0, 2],
    [8, 3, 1]
]

# Calculate trace
result_trace = matrix_trace(example_matrix)

# Display result
print("Original Matrix:")
for row in example_matrix:
    print(row)

print("\nMatrix Trace:")
print(result_trace)
```

\begin{equation}
\begin{split}
A &= \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad B = \begin{bmatrix} -1 & 0 \\ 1 & 1 \end{bmatrix}, \\
2B &= \begin{bmatrix} -2 & 0 \\ 2 & 2 \end{bmatrix}, \quad A - 2B = \begin{bmatrix} 3 & 2 \\ 1 & 2 \end{bmatrix}, \\
X &= \frac{1}{3} \begin{bmatrix} 3 & 2 \\ 1 & 2 \end{bmatrix} = \begin{bmatrix} 1 & \frac{2}{3} \\ \frac{1}{3} & \frac{2}{3} \end{bmatrix}.
\end{split}
\end{equation}

<span style="font-weight: bold;font-size: 16px;">$2(A-B+2X) = 3(X-B)$</span>

\begin{equation}
\begin{split}
2(A - B + 2X) &= 3(X - B), \\
2A - 2B + 4X &= 3X - 3B, \\
2A - 2B - 3X + 4X &= -3B, \\
2A + X - 2B &= -3B, \\
X &= -2A + B.
\end{split}
\end{equation}

Substitute matrices:

$X = -2 \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} + \begin{bmatrix} -1 & 0 \\ 1 & 1 \end{bmatrix}.$


Perform matrix operations:

$X = \begin{bmatrix} -1 & -4 \\ -7 & -9 \end{bmatrix}.$



```python
import numpy as np

# Given matrices A and B
A = np.array([[1, 2], [3, 4]])
B = np.array([[-1, 0], [1, 1]])

# Solve the system of equations 3X = A - 2B
X1 = np.linalg.solve(3 * np.eye(2) - 2 * B, A)

# Solve the system of equations 2(A - B + 2X) = 3(X - B)
X2 = np.linalg.solve(2 * (A - B + 2 * np.eye(2)), 3 * (np.eye(2) - B))

print("Solution for 3X = A - 2B:")
print(X1)

print("\nSolution for 2(A - B + 2X) = 3(X - B):")
print(X2)
```
